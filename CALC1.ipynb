```python

from tkinter import *
import math

root=Tk()
root.title("CALCULATOR")
root.resizable(width=False,height=False)
root.geometry("350x440+600+100")
text_Input=StringVar()
operator=""

def btn(numbers):
    global operator
    operator=operator+str(numbers)
    text_Input.set(operator)
    
def btnClearDisplay():
    global operator
    operator=""
    text_Input.set("")

def btnEqual():
    global operator
    sumup=str(eval(operator))
    text_Input.set(sumup)
    operator=""
       
f2=Frame(root, width=300, height=700, bg="light blue",relief=RAISED)
mainFrame= Frame(root, bd=10, bg='gainsboro', relief=RIDGE )
mainFrame.grid()

f2= Frame(root, bd=5, bg='gainsboro', relief=RIDGE )
f2.grid()

txtDisplay=Entry(f2,font=('arial',20,'bold'),textvariable=text_Input,bd=20,insertwidth=4,bg="powder blue",justify="right")
txtDisplay.grid(columnspan=5)

btn7=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="7",bg="powder blue",command= lambda:btn(7)).grid(row=2,column=0)
btn8=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="8",bg="powder blue",command= lambda:btn(8)).grid(row=2,column=1)
btn9=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="9",bg="powder blue",command= lambda:btn(9)).grid(row=2,column=2)
add=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="+",bg="powder blue",command= lambda:btn("+")).grid(row=2,column=4)
btn4=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="4",bg="powder blue",command= lambda:btn(4)).grid(row=3,column=0)
btn5=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="5",bg="powder blue",command= lambda:btn(5)).grid(row=3,column=1)
btn6=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="6",bg="powder blue",command= lambda:btn(6)).grid(row=3,column=2)
sub=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="-",bg="powder blue",command= lambda:btn("-")).grid(row=3,column=4)
btn1=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="1",bg="powder blue",command= lambda:btn(1)).grid(row=4,column=0)
btn2=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="2",bg="powder blue",command= lambda:btn(2)).grid(row=4,column=1)
btn3=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="3",bg="powder blue",command= lambda:btn(3)).grid(row=4,column=2)
mul=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="*",bg="powder blue",command= lambda:btn("*")).grid(row=4,column=4)
btn0=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="0",bg="powder blue",command= lambda:btn(0)).grid(row=5,column=0)
btnClear=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="C",bg="powder blue",command=btnClearDisplay).grid(row=5,column=1)
btnEquals=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="=",bg="powder blue",command=btnEqual).grid(row=5,column=2)
div=Button(f2,padx=10,pady=10,bd=8,fg="black",font=('arial',20,'bold'),
            text="/",bg="powder blue",command= lambda:btn("/")).grid(row=5,column=4)

root.mainloop()
```
